```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datafile = "LoansTrainingSetV2.csv"
raw_data = pd.read_csv(datafile, header=0)
```

```python
raw_data.head(20)
```

```python
raw_data.describe()
```

```python
# remove the duplicate rows of Loan ID column
df = pd.DataFrame(raw_data)
if df["Loan ID"].duplicated().any():
    df.drop_duplicates(subset=["Loan ID"], keep='first', inplace=True)
df.count()
```

```python
df.dtypes
```

```python
# remove non-digit characters in Monthly Debt column and convert to numeric values
df["Monthly Debt"] = df["Monthly Debt"].replace('[^0-9.]', '', regex=True)
df["Monthly Debt"] = pd.to_numeric(df["Monthly Debt"], errors='coerce')
df["Monthly Debt"].dtypes
```

```python
# convert Maximum Open Credit column to numeric
df["Maximum Open Credit"] = pd.to_numeric(df["Maximum Open Credit"], errors='coerce')
df["Maximum Open Credit"].dtypes
```

```python
# convert Months since last deliquent column to numeric
df["Months since last delinquent"] = pd.to_numeric(df["Months since last delinquent"], errors='coerce')
```

```python
# convert Bankruptcies column to numeric
df["Bankruptcies"] = pd.to_numeric(df["Bankruptcies"], errors='coerce')
```

```python
# convert Tax Liens column to numeric
df["Tax Liens"] = pd.to_numeric(df["Tax Liens"], errors='coerce')
```

```python
# clean missing data in columns Credit Score and Annual Income
df["Credit Score"] = df["Credit Score"].fillna(df["Credit Score"].median())
df["Annual Income"] = df["Annual Income"].fillna(df["Annual Income"].mean())
```

```python
# clip outlier in Credit Score
credit_score_limit = 800
while (any(df["Credit Score"]>credit_score_limit)):
        df.loc[df["Credit Score"]>credit_score_limit, "Credit Score"] = df.loc[df["Credit Score"]>credit_score_limit, "Credit Score"] / 10
```

```python
# Convert string to numeric in column Years in current job
df["Years in current job"] = df["Years in current job"].astype('str')
df.loc[df["Years in current job"]=='< 1 year', "Years in current job"] = '0'
df.loc[df["Years in current job"]=='n/a', "Years in current job"] = '0'
df.loc[df["Years in current job"]=='10+ years', "Years in current job"] = '10'
df["Years in current job"] = df["Years in current job"].replace('[^0-9.]', '', regex=True)
df["Years in current job"] = df["Years in current job"].fillna('0')
```
