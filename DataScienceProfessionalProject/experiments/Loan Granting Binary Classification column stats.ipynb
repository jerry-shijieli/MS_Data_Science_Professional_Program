```python
# convert label to 0 and 1 binary numeric values
#df[df["Loan Status"]=="Charged Off"]["Loan Status"] = '0'
#df[df["Loan Status"]=="Fully Paid"]["Loan Status"] = '1'
df["Loan Status"] = df["Loan Status"].replace("Charged Off", '0')
df["Loan Status"] = df["Loan Status"].replace("Fully Paid", '1')
df["Loan Status"] = df["Loan Status"].astype('int')
df["Loan Status"].unique()
```

```python
import seaborn as sns
sns.set(color_codes=True)
sns_plot = sns.pairplot(df, hue='Loan Status', size=2.5)
sns_plot.savefig('pairplot.png')
```

```python
df.describe()
```

```python
df["Current Loan Amount"] = df["Current Loan Amount"].apply(lambda x: x if x<1e5 else 1e5)
df["Current Loan Amount"].describe()
```

```python
df.hist(column="Years in current job", bins=100)
```

```python
df["Annual Income"] = df["Annual Income"].apply(lambda x: x if x<1e5 else np.nan)
df["Annual Income"].describe()
```

```python
df["Monthly Debt"] = df["Monthly Debt"].apply(lambda x: x if x<15000 else np.nan)
#df["Monthly Debt"] = df["Monthly Debt"].dropna(axis=0)
df["Monthly Debt"].describe()
```

```python
df["Current Credit Balance"] = df["Current Credit Balance"].apply(lambda x: x if x<1e6 else np.nan)
#df["Current Credit Balance"] = df["Current Credit Balance"].dropna(axis=0)
df["Current Credit Balance"].describe()
```

```python
df["Maximum Open Credit"] = df["Maximum Open Credit"].apply(lambda x: np.log(1e-8) if x<=0 else np.log(x))
df["Maximum Open Credit"].describe()
```

```python
df["Term"] = df["Term"].replace("Short Term", '0')
df["Term"] = df["Term"].replace("Long Term", '1')
df["Term"] = df["Term"].astype('int')
df["Term"].unique()
```

```python
df.dropna(axis=0, inplace=True)
df.describe()
```

```python
sns_plot = sns.pairplot(df, hue='Loan Status', size=2.5)
sns_plot.savefig('pairplot_cleaned.png')
```
